```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset (you can replace with CSV later)
data = {
    "Year": [2000, 2005, 2010, 2015, 2020, 2023],
    "Temperature": [14.3, 14.5, 14.7, 14.9, 15.1, 15.3],
    "CO2": [370, 380, 390, 400, 415, 420]
}

df = pd.DataFrame(data)

print("Dataset:\n")
print(df)

# Plot Temperature trend
plt.figure()
plt.plot(df["Year"], df["Temperature"], marker='o')
plt.title("Global Temperature Trend")
plt.xlabel("Year")
plt.ylabel("Temperature (°C)")
plt.show()

# Plot CO2 trend
plt.figure()
plt.plot(df["Year"], df["CO2"], marker='o')
plt.title("CO2 Emissions Trend")
plt.xlabel("Year")
plt.ylabel("CO2 (ppm)")
plt.show()
```
